Ejercicio 3 - A

```python
import pandas as pd
import numpy as np

colm1 = np.linspace(0,4,201)

df = pd.DataFrame(
    colm1, columns = ['t']
)

df
```

Ejercicio 3 - B

```python
from numpy import sin

df['s'] = df['t'].apply(sin)

df
```

Ejercicio 3 - C

```python
df["s10"] = df["s"].shift(10)
df
```

Ejercicio 3 - D

```python
df["d"] = df["s"]-df["s10"]
df
```

Ejercicio 3 - E

```python
def is_positive(x):
  if x > 0 :
    return 1
  else:
    return 0

df["p"] = df["d"].apply(is_positive)
df
```

Ejercicio 3 - E

Ejercicio 6

```python
def generar_primo():
  numero = 2
  yield numero
  while True:
    temp = numero 
    while True:
      temp += 1
      contador = 1
      contador_divisores = 0
      while contador <= temp:
        if temp % contador == 0:
          contador_divisores += 1
        if contador_divisores > 2:
          break
        contador += 1
      if contador_divisores == 2:
        yield temp
        numero = temp
  
g = generar_primo()

primos = [next(g) for _ in range() ]

print(primos)
```
